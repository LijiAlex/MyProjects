### Get meaningful information

How many startups where founded in 2021?

```python
len(funding[funding['Founded']==2021])
```

Maximum amount invested

```python
funding.loc[funding['Amount ($)'].idxmax()]
```

How many times each investor invested

```python
funding['Investor/s'].value_counts()
```

How much each investor invested

```python
InvestorAmount = funding.groupby('Investor/s')['Investor/s','Amount ($)'].sum('Amount ($)')
InvestorAmount.sort_values(by = 'Amount ($)', ascending=False)
```

Count of startups based on location

```python
funding.groupby(['Headquarters']).size()
```

Max and minimum investment based on location for each sectors

```python
groups = funding.groupby(['Headquarters','Sector'])
aggregrate = groups['Amount ($)'].agg([max,min])
```

```python
aggregrate
```

Sector wise investment for a specific location

```python
aggregrate.loc['Gurgaon']
```

```python
aggregrate.loc['Bangalore','Healthcare']
```

Which sector got maximum investment

```python
SectorwiseAmount = funding.groupby('Sector')['Sector','Amount ($)'].max('Amount ($)')
SectorwiseAmount
```

```python
SectorwiseAmount.loc[SectorwiseAmount.idxmax()]
```
